# Students Performance Dataset

```python
student = pd.read_csv('StudentsPerformance.csv')
```

```python
student
```

```python
# 1
```

```python
student['math score'].mean()
```

```python
student['reading score'].mean()
```

```python
student['writing score'].mean()
```

```python
# 2
```

```python
avg_writing_by_gender = student.groupby('gender')['writing score'].mean()
```

```python
avg_writing_by_gender
```

```python
# 3
```

```python
student['total score'] = student[['math score' ,'reading score','writing score']].sum(axis= 1)
```

```python
student['total score']
```

```python
student['percentage'] = ((student['total score'] / 300) * 100).round(2)
```

```python
student
```

```python
# 4
```

```python
student['average score'] = student[['writing score', 'reading score' , 'math score']].mean(axis = 1)
```

```python
avg_by_parent = student.groupby('parental level of education')['average score'].mean().sort_values()

# Step 3: Plot the bar chart
avg_by_parent.plot(kind='barh', color='skyblue', edgecolor='black')

plt.title('Average Score by Parental Level of Education')
plt.xlabel('Average Score')
plt.ylabel('Parental Education Level')
plt.tight_layout()
plt.show()
```
